```python
''' problem statement:
Write a python program to calculate gross salary of employee based on:
If basic <= 10000: HRA = 20%, DA = 80%
If basic <= 20000: HRA = 25%, DA = 90%
Otherwise: HRA = 30%, DA = 95%'''
basic = int(input("enter basic salary: "))
if basic <= 10000:
  HRA = basic*0.2
  DA = basic*0.8
elif basic <= 20000:
  HRA = basic*0.25
  DA = basic*0.90
else :
  HRA = basic*0.3
  DA = basic*0.95
gross = basic + HRA + DA
print("gross salary: ", gross)
```

```python
'''Write a python program to calculate the income tax based on annual income:
Up to $2,50,000: No tax
$2,50,001 - $5,00,000: 5%
$5,00,001 - $10,00,000: 20%
Above $10,00,000: 30%'''
income = int(input("enter your annual income: "))

if income <= 250000:
    tax = 0
elif income <= 500000:
    tax = (income - 250000) * 0.05
elif income <= 1000000:
    tax = (500000 - 250000) * 0.05 + (income - 500000) * 0.20
else:
    tax = (500000 - 250000) * 0.05 + (1000000 - 500000) * 0.20 + (income - 1000000) * 0.30

print("Income Tax: ", tax)
```
