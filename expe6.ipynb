```python
#1. check whether a number is positive, negative or zero
num = int(input("Enter a number: "))
if num > 0:
          print("positive")
elif num<0:
          print("negative")
else:
          print("zero")
```

```python
#2. check if number is even or odd
num = int(input("Enter a number: "))
if num % 2 == 0:
  print("even")
else:
  print("odd")
```

```python
#3. check largest number
a = int(input("Enter a number: "))
b = int(input("Enter b number: "))
c = int(input("Enter c number: "))
if a > b & a > c:
  print("a is largest")
elif b > a and b > c:
  print("b is largest")
else:
  print("c is largest")
  # works only for unequal numbers
```

```python
#4. get input from user for one subject and calculate grade using if-elif-else
marks = float(input("Enter marks of the subject: "))
if marks >= 90:
  print("Grade A")
elif marks >= 75:
  print("Grade B")
elif marks >= 60:
  print("Grade C")
elif marks >= 45:
  print("Grade D")
else:
  print("Grade F")
```

```python
#5. find if given year is leap year or not
year = int(input("Enter a year: "))
if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
  print("leap year")
else:
  print("not a leap year")
```
